```python
import pandas as pd
```

```python
df = pd.read_csv('new_data.csv', names=['Sent', 'Value'])
df['Sent'] = df["Sent"].apply(lambda x: ''.join([" " if ord(i) < 32 or ord(i) > 126 else i for i in x]))
```

```python
df.head()
```

```python
df.dropna(axis=0, how='any')
```

```python
df.dtypes
```

```python
#df['Value'] = df['Value'].str.replace(r'\D+', '')
df['Value'] = df['Value'].astype(int)
```

```python
df.head()
```

```python
df['Sent'] = df['Sent'].astype('str') 
df.dropna(axis=0, how='any')
```

```python
df = df[df['Value'].notnull()]
```

```python
from sklearn.utils import shuffle
df = shuffle(df)
```

```python
len(df
   )
```

```python
from nltk.tokenize import word_tokenize
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['Sent'],df['Value'],test_size=0.33, random_state=53)
count_vec = CountVectorizer(strip_accents=None)
```

```python
count_train = count_vec.fit_transform(X_train, y_train)
```

```python
df.iloc[19]
```

```python
count_test = count_vec.transform(X_test)
#print(count_vec.get_feature_names()[:1000])
```

```python
count_df = pd.DataFrame(count_train.A, columns=count_vec.get_feature_names())
```

```python
count_df.head()
```

```python
count_df = count_df.loc[:,'about':]
```

```python
count_df.head()
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import BernoulliNB

from sklearn import metrics
```

```python
nb_classifier = MultinomialNB()
bernauli_nb = BernoulliNB()
```

```python
nb_classifier.fit(count_df, y_train)
bernauli_nb.fit(count_df, y_train)
```

```python
count_test = pd.DataFrame(count_test.A, columns=count_vec.get_feature_names())
```

```python
count_test = count_test.loc[:,'about':]
```

```python
count_test.head()
```

```python
pred = nb_classifier.predict(count_test)
bernauli_nb.score(count_test, y_test)
```

```python
score = metrics.accuracy_score(y_test, pred)
print(score)
```
